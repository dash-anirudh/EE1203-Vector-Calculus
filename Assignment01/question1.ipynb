### Question 1

$$
\begin{align*}
F(\mathbf{r}) &= -2{\mathbf{r}\over \lVert \mathbf{r} \rVert}\\
&= -2\left({\mathbf{r}\over \sqrt{x^2 + y^2 + z^2}}\right)\\
&= -{2x\over \sqrt{x^2 + y^2 + z^2}}\hat i - {2y\over \sqrt{x^2 + y^2 + z^2}}\hat j - {2z\over\sqrt{x^2 + y^2 + z^2}}\hat k\\
\end{align*}
$$

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')

x, y, z = np.meshgrid(np.arange(-1, 1, 0.15),
                      np.arange(-1, 1, 0.15),
                      np.arange(-1, 1.2, 1))


u = -2 * x / np.sqrt(x**2 + y**2 + z**2)
v = -2 * y / np.sqrt(x**2 + y**2 + z**2)
w = -2 * z / np.sqrt(x**2 + y**2 + z**2)

ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True, 
          color='royalblue', arrow_length_ratio=0.6, 
          linewidth=1.5, alpha=0.8, label=r"${F(\mathbf{r})} = -2{\frac{\mathbf{r}}{\| \mathbf{r}\|}}$")

plt.title("Vector Field Visualisation", fontsize=16)
plt.xlabel(r"$x$", fontsize=12)
plt.ylabel(r"$y$", fontsize=12)
ax.set_zlabel(r"$z$", fontsize=12)

plt.legend(loc="upper right")
plt.savefig("images/question1_a_python_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

$$
\begin{align*}
F(\mathbf{r}) &= 2{\mathbf{e_2}\times \mathbf{r}\over \lVert \mathbf{r} \rVert^2}\\
\end{align*}
$$

First finding the cross product

$$
\begin{align*}
\mathbf{e_2}\times \mathbf{r} &= \begin{vmatrix}
\hat i & \hat j & \hat k\\
0 & 1 & 0\\
x & y & z\\
\end{vmatrix}\\
&= (z - 0)\hat i - (0 - 0)\hat j + (0 - x)\hat k\\
&= z\hat i - x\hat k\\
\end{align*}
$$

Thus, we have

$$
\begin{align*}
F(\mathbf{r}) &= 2\left({\mathbf{e_2}\times \mathbf{r}\over \lVert \mathbf{r} \rVert^2}\right)\\
&= 2\left({z\hat i - x\hat k\over x^2 + y^2 + z^2}\right)\\
&= {2z\over x^2 + y^2 + z^2}\hat i + 0\hat j -{2x\over x^2 + y^2 + z^2}\hat k\\
\end{align*}
$$

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')

x, y, z = np.meshgrid(np.arange(-1, 1, 0.15),
                      np.arange(-1, 1, 0.15),
                      np.arange(-1, 1.2, 1))


u = 2*z/(x**2 + y**2 + z**2)
v = 0
w = -2*x/(x**2 + y**2 + z**2)

ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True, 
          color='royalblue', arrow_length_ratio=0.6, 
          linewidth=1.5, alpha=0.8, label=r"$\mathbf{F(r)} = -2{\frac{\mathbf{e_2 \times r}}{\| \mathbf{r}\|^2}}$")

plt.title("Vector Field Visualization", fontsize=20)
plt.xlabel(r"$x$", fontsize=16)
plt.ylabel(r"$y$", fontsize=16)
ax.set_zlabel(r"$z$", fontsize=16)

plt.legend(loc="upper right")
plt.savefig("images/question1_b_python_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```
